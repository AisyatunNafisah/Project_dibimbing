4. mengatasi missing value

```python
# Mengatasi missing value
for column in df.columns:
    if df[column].isnull().any(): # Penting: Hanya proses kolom yang memiliki NaN
        if df[column].dtype == 'object':
            # Jika kolom bertipe object, isi dengan mode dan assign kembali
            df[column] = df[column].fillna(df[column].mode()[0])
            print(f"  --> Kolom '{column}' (object): NaN diisi dengan modus '{df[column].mode()[0]}'")
        else:
            # Jika kolom bertipe numerik, isi dengan mean dan assign kembali
            # Menggunakan pd.api.types.is_numeric_dtype untuk cek yang lebih robust
            if pd.api.types.is_numeric_dtype(df[column]):
                mean_val = df[column].mean()
                df[column] = df[column].fillna(mean_val)
                print(f"  --> Kolom '{column}' (numeric): NaN diisi dengan rata-rata '{mean_val:.2f}'")
            else:
                print(f"  --> Kolom '{column}': Tipe data tidak numerik, tidak diisi.")
    else:
        print(f"  --> Kolom '{column}': Tidak ada NaN.")

print("\n--- Proses Imputasi Selesai ---")
# Anda bisa cek df.info() atau df.isnull().sum() lagi setelah ini
```
